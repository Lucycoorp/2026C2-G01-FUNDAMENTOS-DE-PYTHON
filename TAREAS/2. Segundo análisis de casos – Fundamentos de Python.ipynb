# Problema #1: Calculadora de propina inteligente

## Problema

EL Programa Solicita el total de la cuenta de restaurante y porcentaje de porpina que eñl usuario desea dejar con esos dos datos calcula el monto de la porpina y total final a pagar 

## Entrada

 - Total de la cuenta (numero decimal. Ej. 15000)
 - Porcentaje de porpiina deseado (numero decimal, Ej. 10)


## Proceso

- propina = total * (porcentaje / 100)
- total_final = total + propina

## Salida esperada

- Monto de la propina  calculada
- Total final a pagar

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|---------------------|
| Caso normal | total=15000, propina=10% | Propina: ₡1500.00 / Total: ₡16500.00 | Propina: ₡1500.00 / Total: ₡16500.00 |
| Caso límite | total=0, propina=0% | Propina: ₡0.00 / Total: ₡0.00 | Propina: ₡0.00 / Total: ₡0.00 |
| Caso inválido | total="abc" | Error: ingrese un número válido. | Error: ingrese un número válido. |


```python
# Escriba la solución aquí
try:
    total = float(input("ingreseel total de la cuenta"))
    porcentaje = float (input("Ingrese el porcentaje de propina (%):"))
    Propina = total * (porcentaje / 100)
    total_final = total + Propina
    print(f"Propina ({porcentaje}%): ₡ {Propina:.2f}")
    print (f"Total a pagar:          ₡{total_final:.2f}")
except ValueError:
    print("Error: ingrese un numero válido")
    
```

# Problema #2: Detector de número par/impar

## Problema

El Programa Solicita un numero entero y determina si es par o impar.


## Entrada

- un numero entero (ej.7)

## Proceso

- Si numero % 2 == 0 → par
- De contario  → impar

## Salida esperada

- Mensaje que indica si el numero es par o impar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|---------------------|
| Caso normal | 7 | 7 es impar. | 7 es impar. |
| Caso límite | 0 | 0 es par. | 0 es par. |
| Caso inválido | "hola" | Error: ingrese un número entero válido. | Error: ingrese un número entero válido. |


```python
# Escriba la solución aquí
try:
    numero = int(input("Ingrese un numero entero: "))
    if numero % 2 == 0:
        print(f"{numero}es par.")
    else:
        print (f"{numero}es impar.")
except ValueError: 
       print("error: ingrese un numero entero valido.")
```
