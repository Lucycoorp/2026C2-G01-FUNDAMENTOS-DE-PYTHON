# Problema #1: Calculadora de propina inteligente

## Problema

EL Programa Solicita el total de la cuenta de restaurante y porcentaje de porpina que eñl usuario desea dejar con esos dos datos calcula el monto de la porpina y total final a pagar 

## Entrada

 - Total de la cuenta (numero decimal. Ej. 15000)
 - Porcentaje de porpiina deseado (numero decimal, Ej. 10)


## Proceso

- propina = total * (porcentaje / 100)
- total_final = total + propina

## Salida esperada

- Monto de la propina  calculada
- Total final a pagar

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|---------------------|
| Caso normal | total=15000, propina=10% | Propina: ₡1500.00 / Total: ₡16500.00 | Propina: ₡1500.00 / Total: ₡16500.00 |
| Caso límite | total=0, propina=0% | Propina: ₡0.00 / Total: ₡0.00 | Propina: ₡0.00 / Total: ₡0.00 |
| Caso inválido | total="abc" | Error: ingrese un número válido. | Error: ingrese un número válido. |


```python
# Escriba la solución aquí
try:
    total = float(input("ingreseel total de la cuenta"))
    porcentaje = float (input("Ingrese el porcentaje de propina (%):"))
    Propina = total * (porcentaje / 100)
    total_final = total + Propina
    print(f"Propina ({porcentaje}%): ₡ {Propina:.2f}")
    print (f"Total a pagar:          ₡{total_final:.2f}")
except ValueError:
    print("Error: ingrese un numero válido")
    
```

# Problema #2: Detector de número par/impar

## Problema

El Programa Solicita un numero entero y determina si es par o impar.


## Entrada

- un numero entero (ej.7)

## Proceso

- Si numero % 2 == 0 → par
- De contario  → impar

## Salida esperada

- Mensaje que indica si el numero es par o impar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|---------------------|
| Caso normal | 7 | 7 es impar. | 7 es impar. |
| Caso límite | 0 | 0 es par. | 0 es par. |
| Caso inválido | "hola" | Error: ingrese un número entero válido. | Error: ingrese un número entero válido. |


```python
# Escriba la solución aquí
try:
    numero = int(input("Ingrese un numero entero: "))
    if numero % 2 == 0:
        print(f"{numero}es par.")
    else:
        print (f"{numero}es impar.")
except ValueError: 
       print("error: ingrese un numero entero valido.")
```

-

# Problema #4: Validador de compra con descuento

## Problema

El problema solicita el nombre del cliente, el monto de la compra y si es cliente frecuenta. Aplica un descuento de 15 % si el cliente frecuenta y el monto supera ₡10 000;
un 5% si es cliente frecuente con un monto igual o menor, sin descuento en cualquier oto caso.

## Entrada

- Nombre del cliente (texto)
-Monto de la compra (numero decimal)
si es cliente frecuente (s/n)


## Proceso

- frecuente == "s"` y monto > 10000 → descuento 15 %.
- frecuente == "s"` y monto <= 10000 → descuento 5 %.
- frecuente == "n"` → sin descuento.
- total = monto - (monto * descuento)


## Salida esperada

- Nombre del cliente, descuento aplicado y total a pagar.


## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|---------------------|
| Caso normal | Ana, 20000, s | Descuento 15% / Total: ₡17000.00 | Descuento 15% / Total: ₡17000.00 |
| Caso límite | Luis, 10000, s | Descuento 5% / Total: ₡9500.00 | Descuento 5% / Total: ₡9500.00 |
| Caso inválido | María, "abc", s | Error: ingrese un monto válido. | Error: ingrese un monto válido. |


```python
# Escriba la solución aquí
try:
    cliente = input("Nombre del cliente: ").strip()
    monto = float(input("Monto de la compra (₡): "))
    Frecuente = input("¿Es cliente frecuente? (s/n): ").strip().lower()
    
    if Frecuente == "s" and monto > 10000:
        descuento  = 0.15
    elif Frecuente == "s":
        descuento = 0.05
    else: 
        descuento = 0.0
    
    ahorro = monto * descuento 
    total = monto - ahorro
    
    print(f"cliente:        {cliente}")
    print(f"Monto original ₡{monto:.2f}")
    print(f"Descuento ({int(descuento*100)}%):    ₡{ahorro:.2f}")
    print(f"total a pagar:  ₡{total:.2f}")
    
except ValueError:
    print ("Error: ingrese un monto valido") 
        
```
